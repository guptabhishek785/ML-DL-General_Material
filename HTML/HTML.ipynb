HTML - Hypertext Markup Language

- HTML is the code that is used to give a structure to a webpage and its content
- Component used to design the structure of websites are called HTML tags

Note:

- server by defaults considers index.html as its home page file and opens it first. If we want another file to be the homepage then we need to instruct manually. Thus prefer index.html to be the home page

HTML Tags: Container for content or some other html tags


```
<p> Hello </p>
```

- The entire above line is one element
- `<p>` is the opening tag
- `</p>` is the closing tag
- Middle portion is the content

- Head tag - contains that that is not displayed on the webpage but gives information to the browser about the wepage i.e. metadata
e.g. viewport, title, etc

- Body tag - contains all the information that is to be displayed on the webpage

## HTML Attributes: used to add more information to the tag

1. Heading tag: used to display headings in html
e.g. h1, h2,..., h6     (decreasing level of importance)

Note: Headings should be used wrt to the importance and not wrt the size needed as it can be changed using attributes or css. But the importance is used in SEO for ranking the pages and thus headings should be used accordingly

2. Paragrah tag: used to add paragraphs in html                 
e.g. `<p> This is a paragraph </p>`

3. Anchor tag: used to add links to our webpage                          
e.g.` <a href = "link" > content < /a>`
- Anchor tag is also used to cross links other webpages as well

4. Image tag: used to add images                
e.g. `<img src = "/notebook_img.jpg" alt = "notebook"></>`
- if the image does not get printed due to any reason than the "alt" content is displayed
- height and width attributes are used to resize the image. Generally it is advisable to resize either height or width not both at once

5. Next Line: used to add next line/ line breaks
- br tag is en empty tag that it does not have additional attributes

6. Bold, italic & underline: used to highlight a tag

```
<b> Bold </b>
<i> italic </i>
<u> underline </u>
```

7. Big and Small tag: Used to print Big and Small tag
```
<big>This is Big tag</big>
<br>
<small>This is Small tag</small>
```
8. Hr tag: used to display a horizontal ruler, used to seperate content        
`<Hr>`

9. Subscript and Superscript tag
```
<Hr>
<p>H<sub>2</sub>O</p>
<Hr>
<p>x<sup>2</sup></p>
```

## Page Layout Tags:
Semantic tags: tags which give the literal meaning from their names e.g. headers. footers
Semantic tags help in code readability and also helps in SEO rankings

- Section Tag: For a section on your page
- Article Tag: For an article on your page
- Aside Tag: For content aside main content(e.g. for ads, SEO ignores aside tag)

Non Semantic tags: which do not give meaning from their names e.g. div, span

- Div tags: Div is a container used for other HTML elements. Div is a block element (takes full width). if div is empty its height is zero

- Span tags: Span is also a container used for other HTML elements. Span is an inline elements (takes minimal width)

-Div and span are mainly used when multiple elements are to be changed at once and also for consitency

## Forms: forms are used to collect data from the user
e.g. login, contact, signup

```
<form>
Form content
</form>
```

Additioinal attributes:
- action: The action to be performed once the form is submitted
- inputs: type (text, password(displays bullets), placeholder (temporary text written)
- label: defines a label for form attributes. It is mainly used to improve user experience. It highlights when  cursor hovers over and screen-reader reads out loud the label. also pressing on the label means pressing on the radio buton
- name in the radio button ensures that only one button is pressed for all elements with the same name
- Value is what goes to the backend. Text is not sent to the backend. It is just for the user


- Class: Class is assigning a single group to the elements
- id: id is uniqu for each element and helps differentiate similar elements. eg in a class
