## Lecture 4: Feature Engineering

Feature Engineering is the process of converting data i.e. (text, images, voice) into vectors/ numbers to feed into the ML algo.

*This is one of the most important steps in ML*

#### Important Terms in NLP

- Corpus
- Vocabulary
- Document
- Word

- Out of Vocabulary (OOV): Words that were not present in training data but appear in test data

#### Techniques of Text Vectorization

- One Hot Encoding
  - Not used in Industry. No library to implement on code
  - Disadvantages:
    1. Very Very Sparse. Consumes lot of memory since each document is a 3D vector. Also document does not have a fixed size.
    2. Does not address Out of Vocabulary problems (OOV)
    3. No capturing of Semantic Meaning
    4. Does not consider order of words. Also combination of words is not considered

- Bag of Words. Library = sklearn.feature_extraction.text.CountVectorizer
  - Same problems as above. Except it has a fixed array size. Each document is a 1D array with size = size of Vocabulary
  - In CountVectorizer Lib (imp Hyperparameters):
      binary = True: used when dealing with Sentiment Analysis
      Max features: used when we have a very large value of v, but a lot of the words do not occur often but increase the size

- ngrams
  - Same as BOW except it considers a group of words as one elemnt of the vector. BOW is special case of ngrams where n = 1. Library is same i.e. Count Vectorizer where ngrams = (2, 2)
  Value of ngrams is changed to make some combination i.e. (1, 2) or (1,3)
  - Disadvantages:
  1. Able to capture some Semantic Meaning but not entirely
  2. Easy to implement
  3. Slows down algo since as n increases sparsity also increases
  4. Rest same as BOW

- tfidf. Library = tfidfVectorizer
  Problem with n-grams is that certain words which appear frequently in the document gain more importance. This gets worsened when there are lot of frequent unimportant words. Instead more importance is needed to be given to words that appear less frequently since they bring new value in the text. This is achieved by tf-idf.

## Lecture 5: Word Embeddings

Word Embeddings: term used for representation of words for text analysis, typically in the form of real-valued vector that encodes the meaning of the word such that the words that are closer in the vector space are expected to be similar in meaning

Types of Word Embeddings
1. Frequency based
  - BOW, tfidf, GLOVE
2. predictoin based
  - Word2Vec


- Word2Vec
  - Technique created by Google Engineers
  - It captures semantic meaning i.e. happy and joy vectors are similar
  - Low Dimenision vectors i.e 300 X 1
  - This is because vectors are dense and not sparse
  - Can be pretrained or self trained
  - Gensim is one of the pretrained models with vectors for 300 million words from Google news. Nos in array range from -3 to 3

  - Gensim word2vec also supports vector arithmetic i.e. king - man + woman = queen
  - The underlying assumption of word2vec is that two words sharing similar contexts also share a similar meaning and consequently a similar vector representationo from the model
- 2 types of word2vec based on different NN architecture
  1. CBOW i.e. Continuos Bag of words
  2. skip-gram

Basic idea of CBOw is that given the context words that is the words appearing in a moving window i.e. before the target word and after, we try to predict the target word. DL methods are used to extract weights that would best predict the target words. The weights together form the vector od the word

skip-gram is the inverse of CBOW in the sense that given target word,  the task is to predict the context words.The DL method to extract weights is similar.

- When working with smaller data it is advisable to use CBOW whereas for larger data skip-gram should be preferred

  To improve the accuracy of word2vec
    1. increase the training data
    2. increase dimensions of the vectors
    3. increase window size
